```python
number = int(input("Please enter the number - "))

sum_digits = 0

while number > 0:
    last_digit = number % 10
    sum_digits += last_digit
    number = number // 10

print("Sum of individual digits =", sum_digits)
```

```python
'''
Build a meter system for an auto. Take into two inputs. Kilometers Travelled and Stall Time (in minutes). Rate List -
First 10 kms - Rs 10 per Km
Next 40 kms - Rs 9 per Km
Next 100 kms - Rs 8 per Km
Any leftover km count - Rs. 6 per Km Rs 5 extra for every minute of Stall Time
'''
```

```python
kilometers = float(input())
minutes = float(input())

rate = 0

if kilometers > 150:
    rate += (kilometers - 150)*6 + 10*10 + 40*9 + 100*8
elif kilometers > 50:
    rate += (kilometers - 50)*8 + 10*10 + 40*9
elif kilometers > 10:
    rate += (kilometers - 10)*9 + 10*10
else:
    rate += kilometers*10
    
if minutes:
    rate += minutes*5
    
print("Final Fare -", rate)
```
